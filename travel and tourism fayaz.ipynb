```python
import pandas as pd
```

```python
df = pd.read_csv("tourism_dataset.csv")
```

```python
print(df.head())
```

```python
category_col = df["Category"]
print(category_col.head())
```

```python
subset_cols = df[["Country", "Visitors", "Rating"]]
print(subset_cols.head())
```

```python
print(df["Visitors"].mean())  
print(df["Revenue"].max()) 
```

```python
filtered_data = df[df["Rating"] > 4]
print(filtered_data)
```

```python
df["Revenue_Per_Visitor"] = df["Revenue"] / df["Visitors"]
print(df.head())
```

```python
df_dropped = df.drop(["Accommodation_Available"], axis=1)
print(df_dropped.head())
```

```python
sorted_data = df.sort_values(by="Visitors", ascending=False)
print(sorted_data.head())
```

```python
import matplotlib.pyplot as plt
```

```python
plt.plot(df["Visitors"][:10], df["Rating"][:10])
plt.xlabel("Visitors")
plt.ylabel("Rating")
plt.title("Visitors vs Rating")
plt.show()
```

```python
countries = df["Country"][:5]
revenues = df["Revenue"][:5]
plt.bar(countries, revenues, color="green")
plt.xlabel("Country")
plt.ylabel("Revenue")
plt.title("Revenue by Country")
plt.show()
```

```python
plt.hist(df["Rating"], bins=10, color="blue", alpha=0.7)
plt.xlabel("Ratings")
plt.ylabel("Frequency")
plt.title("Distribution of Ratings")
plt.show()
```

```python
plt.scatter(df["Visitors"], df["Revenue"], alpha=0.5, color="red")
plt.xlabel("Visitors")
plt.ylabel("Revenue")
plt.title("Visitors vs Revenue")
plt.show()
```

```python
categories = df["Category"].value_counts()
plt.pie(categories, labels=categories.index, autopct="%1.1f%%", startangle=140)
plt.title("Distribution of Categories")
plt.show()
```
